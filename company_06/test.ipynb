```python
historical_data["date"] = historical_data["date"].astype(str)
historical_data[["year", "month"]] = historical_data["date"].str.split("-", expand=True)
```

```python
# Mapping of numeric months to Nepali month names
nepali_months = {
    "01": "Baisakh", "02": "Jestha", "03": "Ashadh",
    "04": "Shrawan", "05": "Bhadra", "06": "Ashoj",
    "07": "Kartik", "08": "Mangsir", "09": "Poush",
    "10": "Magh", "11": "Falgun", "12": "Chaitra"
}

# Apply mapping to replace month numbers with names
historical_data["nepali_date"] = historical_data["year"] + "-" + historical_data["month"].map(nepali_months)

print(historical_data[["nepali_date"]])
```
